```python
def total_effectiveness(type1, type2):
    effectiveness = 1.0
    if type1 in type_effectiveness and type2 in type_effectiveness:
        effectiveness *= type_effectiveness[type1][type2]
        effectiveness *= type_effectiveness[type2][type1]
    return effectiveness

df['Total Effectiveness'] = df.apply(lambda row: total_effectiveness(row['Type 1'], row['Type 2']), axis=1)

types = list(type_effectiveness.keys())
effectiveness_matrix = [[type_effectiveness[type1][type2] for type2 in types] for type1 in types]
plt.figure(figsize=(10, 8))
sns.heatmap(effectiveness_matrix, cmap='coolwarm', annot=True, fmt=".1f", xticklabels=types, yticklabels=types)
plt.xlabel('Defending Type')
plt.ylabel('Attacking Type')
plt.title('Type Effectiveness Matrix')
plt.tight_layout()
```
