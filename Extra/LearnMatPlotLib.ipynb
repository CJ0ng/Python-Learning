```python
import matplotlib.pyplot as plt
%matplotlib inline
# To see the graph in the notebook

import numpy as np
import pandas as pd
```

```python
# Basic Plotting
x = list(range(0,10))
y = list(range(-10,0))

plt.plot(x,y)
```

```python
#Graph Label

a = [ 100, 86, 1, 59, 32]
b = [ -123, 52, 348, -45, 6]

plt.title("Random Plotting")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(a,b)
```

```python
# Histogram
f = [32,2,1,2,56]
g = [ 42,67,11,2,2]
plt.hist (f)
plt.hist (g)
```

```python
# Functional Plot

x_1 = np.linspace(0,5,10)
y_1 = x_1**2
plt.plot(x_1,y_1)
plt.title('Days Squared Chart')
plt.xlabel('Days')
plt.ylabel('Days Squared')
```

```python
# Print Multiple Plot  ---   first one is split to how many row, 2nd is split to how many column, 3rd is the placement
plt.subplot(1,3,1) 
plt.plot(x_1,y_1,'r')

plt.subplot(1,3,2)
plt.plot(x_1,y_1,'b')

plt.subplot(2,3,3)
plt.plot(x_1,y_1,'g')

plt.subplot(2,3,6)
plt.plot(x_1,y_1,'y')
```

```python
# Figure Objects -- contain all the plot element and can contain many axis

fig1 = plt.figure(figsize = (5,4),dpi = 100)
axes1 = fig1.add_axes([0.1,0.1,0.9,0.9])
axes1.set_xlabel('Days')
axes1.set_ylabel('Days Squared')
axes1.set_title(' Days Squared Chart')
axes1.plot(x_1,y_1,label = 'x/x2')
axes1.plot(y_1,x_1,label = 'x2/x')
axes1.legend(loc=0)
#legened 0 means the system decide where is best ( diff number diff location 1-4)

axes2 = fig1.add_axes([0.45,0.45,0.4,0.3])
axes2.set_xlabel('Days')
axes2.set_ylabel('Days Squared')
axes2.set_title(' Days Squared Chart')
axes2.plot(y_1,x_1,'r')
axes2.legend(loc=0)
```
